```python
import pandas as pd
import numpy as np
import os

os.chdir("J:\MTech Notes\FOML\Assignments\Kaggle challenge")
df=pd.read_csv("train.csv",low_memory=False)

#Removing irrelevant columns

unwanted_cols=["Report Number","Local Case Number",'Crash Date/Time',"Person ID","Vehicle ID","Cross-Street Name",
              "Location","Off-Road Description","Municipality","Non-Motorist Substance Abuse","Related Non-Motorist",
              'Vehicle Make', 'Vehicle Model','Road Name','Agency Name','Circumstance']

df.drop(unwanted_cols, axis=1, inplace=True)

nan_values = df.isna()
nan_columns = nan_values.any()

columns_with_nan = df.columns[nan_columns].tolist()

#null value replacement

df=df.apply(lambda x:x.fillna(x.mean()) if x.dtype=='float' else 
                x.fillna(x.value_counts().index[0]))

df=df.rename(columns={'Route Type': 'Route_Type'})
df['Route_Type'] = df['Route_Type'].replace(np.nan, 0) 

df=df.rename(columns={'ACRS Report Type': 'ACRS_Report_Type'})
labels = df['ACRS_Report_Type'].astype('category').cat.categories.tolist()
replace_map_comp = {'ACRS_Report_Type' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_map_comp, inplace=True)

df=df.rename(columns={'Drivers License State': 'Drivers_License_State'})
labels = df['Drivers_License_State'].astype('category').cat.categories.tolist()
replace_map_comp = {'Drivers_License_State' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_map_comp, inplace=True)

labels = df['Route_Type'].astype('category').cat.categories.tolist()
replace_map_comp = {'Route_Type' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_map_comp, inplace=True)

df=df.rename(columns={'Cross-Street Type': 'Cross_Street_Type'})
labels = df['Cross_Street_Type'].astype('category').cat.categories.tolist()
replace_map_comp = {'Cross_Street_Type' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_map_comp, inplace=True)

df=df.rename(columns={'Collision Type': 'Collision_Type'})
labels = df['Collision_Type'].astype('category').cat.categories.tolist()
replace_map_comp = {'Collision_Type' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_map_comp, inplace=True)

df[['Weather']]=df[['Weather']].replace(dict.fromkeys(['BLOWING SAND, SOIL, DIRT'], 6))  
df[['Weather']]=df[['Weather']].replace(dict.fromkeys(['RAINING','CLOUDY'], 1))
df[['Weather']]=df[['Weather']].replace(dict.fromkeys(['WINTRY MIX','BLOWING SNOW','SNOW'], 2))
df[['Weather']]=df[['Weather']].replace(dict.fromkeys(['FOGGY'], 3))
df[['Weather']]=df[['Weather']].replace(dict.fromkeys(['SLEET'], 5))
df[['Weather']]=df[['Weather']].replace(dict.fromkeys(['UNKNOWN','CLEAR'], 0))
df[['Weather']]=df[['Weather']].replace(dict.fromkeys(['OTHER','SEVERE WINDS'], 12))
df['Weather'] = df['Weather'].replace(np.nan, 0) 

df=df.rename(columns={'Surface Condition': 'Surface_Condition'})      
df[['Surface_Condition']]=df[['Surface_Condition']].replace(dict.fromkeys(['WET','WATER(STANDING/MOVING)'], 1))
df[['Surface_Condition']]=df[['Surface_Condition']].replace(dict.fromkeys(['SNOW','ICE','SLUSH'], 2))
df[['Surface_Condition']]=df[['Surface_Condition']].replace(dict.fromkeys(['SAND','MUD, DIRT, GRAVEL'], 6))
df[['Surface_Condition']]=df[['Surface_Condition']].replace(dict.fromkeys(['OIL', 'UNKNOWN','OTHER','DRY'], 0))
df['Surface_Condition'] = df['Surface_Condition'].replace(np.nan, 0)


labels = df['Light'].astype('category').cat.categories.tolist()
replace_map_comp = {'Light' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_map_comp, inplace=True)


df=df.rename(columns={'Traffic Control': 'Traffic_Control'}) 
labels = df['Traffic_Control'].astype('category').cat.categories.tolist()
replace_map_comp = {'Traffic_Control' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_map_comp, inplace=True)

df=df.rename(columns={'Driver Substance Abuse': 'Driver_Substance_Abuse'}) 
df[['Driver_Substance_Abuse']]=df[['Driver_Substance_Abuse']].replace(dict.fromkeys(['NONE DETECTED'], 5))
df[['Driver_Substance_Abuse']]=df[['Driver_Substance_Abuse']].replace(dict.fromkeys(['MEDICATION CONTRIBUTED','MEDICATION PRESENT'],1))
df[['Driver_Substance_Abuse']]=df[['Driver_Substance_Abuse']].replace(dict.fromkeys(['ALCOHOL CONTRIBUTED','ALCOHOL PRESENT'],2))
df[['Driver_Substance_Abuse']]=df[['Driver_Substance_Abuse']].replace(dict.fromkeys(['ILLEGAL DRUG PRESENT','ILLEGAL DRUG CONTRIBUTED'],3))
df[['Driver_Substance_Abuse']]=df[['Driver_Substance_Abuse']].replace(dict.fromkeys(['COMBINATION CONTRIBUTED','COMBINED SUBSTANCE PRESENT'],4))
df[['Driver_Substance_Abuse']]=df[['Driver_Substance_Abuse']].replace(dict.fromkeys(['UNKNOWN','OTHER'],6))

df=df.rename(columns={'Injury Severity': 'Injury_Severity'}) 
labels = df['Injury_Severity'].astype('category').cat.categories.tolist()
replace_map_comp = {'Injury_Severity' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_map_comp, inplace=True)

df=df.rename(columns={'Vehicle Damage Extent': 'Vehicle_Damage_Extent'})
labels = df['Vehicle_Damage_Extent'].astype('category').cat.categories.tolist()
replace_inju = {'Vehicle_Damage_Extent' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_inju, inplace=True)

df=df.rename(columns={'Vehicle First Impact Location': 'Vehicle_First_Impact_Location'})
df=df.rename(columns={'Vehicle Second Impact Location': 'Vehicle_Second_Impact_Location'})


labels = df['Vehicle_First_Impact_Location'].astype('category').cat.categories.tolist()
replace_inju = {'Vehicle_First_Impact_Location' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_inju, inplace=True)
labels = df['Vehicle_Second_Impact_Location'].astype('category').cat.categories.tolist()
replace_inju = {'Vehicle_Second_Impact_Location' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_inju, inplace=True)

df=df.rename(columns={'Vehicle Body Type': 'Vehicle_Body_Type'})
labels = df['Vehicle_Body_Type'].astype('category').cat.categories.tolist()
replace_inju = {'Vehicle_Body_Type' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_inju, inplace=True)

df=df.rename(columns={'Vehicle Movement': 'Vehicle_Movement'})
labels = df['Vehicle_Movement'].astype('category').cat.categories.tolist()
replace_inju = {'Vehicle_Movement' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_inju, inplace=True)


df=df.rename(columns={'Vehicle Continuing Dir': 'Vehicle_Continuing_Dir'})
df=df.rename(columns={'Vehicle Going Dir': 'Vehicle_Going_Dir'})
labels = df['Vehicle_Continuing_Dir'].astype('category').cat.categories.tolist()
replace_inju = {'Vehicle_Continuing_Dir' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_inju, inplace=True)


labels = df['Vehicle_Going_Dir'].astype('category').cat.categories.tolist()
replace_inju = {'Vehicle_Going_Dir' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_inju, inplace=True)

df=df.rename(columns={'Driverless Vehicle': 'Driverless_Vehicle'})
df=df.rename(columns={'Parked Vehicle': 'Parked_Vehicle'})

labels = df['Driverless_Vehicle'].astype('category').cat.categories.tolist()
replace_inju = {'Driverless_Vehicle' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_inju, inplace=True)

labels = df['Parked_Vehicle'].astype('category').cat.categories.tolist()
replace_inju = {'Parked_Vehicle' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_inju, inplace=True)

df=df.rename(columns={'Equipment Problems': 'Equipment_Problems'})
labels = df['Equipment_Problems'].astype('category').cat.categories.tolist()
replace_inju = {'Equipment_Problems' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
df.replace(replace_inju, inplace=True)

df=df.rename(columns={'Vehicle Year': 'Vehicle_Year'}) 

#grouping the vehicle year as old(1) and new(2) and others under 0
for i in range(0,1801):
    df[['Vehicle_Year']]=df[['Vehicle_Year']].replace(dict.fromkeys([i], 0))
for i in range(1800,2001):
    df[['Vehicle_Year']]=df[['Vehicle_Year']].replace(dict.fromkeys([i], 1))

for i in range(2000,2022):
    df[['Vehicle_Year']]=df[['Vehicle_Year']].replace(dict.fromkeys([i], 2)) 
for i in range(2022,10000):
    df[['Vehicle_Year']]=df[['Vehicle_Year']].replace(dict.fromkeys([i], 0))    

df=df.rename(columns={'Speed Limit': 'Speed_Limit'}) 
df['Speed_Limit'] = np.where(df['Speed_Limit'].between(0,40), 0, df['Speed_Limit'])
df['Speed_Limit'] = np.where(df['Speed_Limit'].between(40,100), 1, df['Speed_Limit'])

#data pre-processing ends here

from sklearn.model_selection import KFold 
from sklearn.metrics import accuracy_score
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
X, y = df.drop('Fault', axis = 1), df['Fault']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=32)




```

```python
##Model 1: XG Boost Classifier
```

```python
%%time
model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', n_estimators=910,learning_rate=0.091,
                      importance_type='gain',random_state=32)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Model 2 Lightgbm Classifier
```

```python
%%time

import lightgbm as ltb
model = ltb.LGBMClassifier(boosting_type='gbdt',max_depth=13,num_leaves=32,
        learning_rate=0.091,n_estimators=895, random_state=32, reg_alpha=0.4, 
        reg_lambda=0.6)
        
model.fit(X_train, y_train)

predicted_y = model.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, predicted_y))
```

```python
#tuning parameters of Light gbm. Final model is the above one

import lightgbm as ltb

start =0.1
stop = 1
step = 0.1

float_range_array = np.arange(start, stop, step)

for i in float_range_array:
    model = ltb.LGBMClassifier(boosting_type='gbdt',max_depth=13,num_leaves=32,
        learning_rate=i,n_estimators=895, random_state=32, reg_alpha=0.4, 
        reg_lambda=0.6)
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)

    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
